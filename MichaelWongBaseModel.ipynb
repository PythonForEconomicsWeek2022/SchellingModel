```python
import matplotlib.pyplot as plt
import itertools
import random
import copy
import pandas as pd
import numpy as np
```

```python
class Schelling:
    def __init__(self,width,height,empty_ratio,similarity_threshold,n_iterations,races):
        self.width = width
        self.height = height
        self.empty_ratio = empty_ratio
        self.similarity_threshold = similarity_threshold
        self.n_iterations = n_iterations
        self.races = races
        
    def populate(self):
        self.agents = {}
        self.all_houses = list(itertools.product(list(range(0,self.width)), list(range(0,self.height))))
        random.shuffle(self.all_houses)
        
        self.n_empty = int(self.width*self.height*self.empty_ratio)
        self.empty_houses = self.all_houses[-self.n_empty:]
        self.remaining_houses = self.all_houses[:-self.n_empty]
        
        for agent in self.remaining_houses:
            self.agents[agent] = random.randint(1,self.races)
    
    def plot_city(self):
        colour = {1:'r',2:'b',3:'g'}
        fig, ax = plt.subplots()
        for key,value in self.agents.items():
            ax.scatter(key[0]+0.5,key[1]+0.5, marker = 's',s=500,c=colour[value])
        plt.subplots_adjust(left=0, right=0.65, bottom=0,top=0.83)
        plt.grid()
        plt.xticks(list(range(self.width)))
        plt.yticks(list(range(self.height)))
        plt.show()
        
    def is_unsatisfied(self, coord, race): #check if neighbours are the same race
        similarity  = 0
        n_neighbours = 0
        all_neighbours = list(itertools.product(list(range(-1,2)), list(range(-1,2))))
        all_neighbours = [(all_neighbours[i][0]+coord[0],all_neighbours[i][1]+coord[1]) for i in range(len(all_neighbours))]
        all_neighbours.remove(coord)
        
        for neighbour in all_neighbours:
            try:
                if self.agents[neighbour] == race:
                    similarity += 1
                    n_neighbours += 1
                elif self.agents[neighbour] != race:
                    n_neighbours += 1
            except:
                pass
        try:
            similarity_index = similarity/n_neighbours
        except:
            similarity_index = 0
        unsatisfied = similarity_index < self.similarity_threshold
        
        return unsatisfied, similarity_index
            
    def update(self):
        similar_list = []
        for key,value in self.agents.copy().items():
            unsatisfied, similarity_index = self.is_unsatisfied(key,value)
            similar_list.append(similarity_index)
            
            if unsatisfied:
                new_key = random.choice(self.empty_houses)
                self.agents[new_key] = self.agents.pop(key)
                self.empty_houses[self.empty_houses.index(new_key)] = key
        
        self.total_s_index = sum(similar_list)/(self.width*self.height)
        print(self.total_s_index)

    def run_sim(self):
        self.populate()
        for iterate in range(self.n_iterations):
            self.plot_city()
            self.update()
            
    def run_sim_until_similar(self):
        self.populate()
        self.plot_city()
        self.update()
        no_iterations = 1
        while True:
            if self.total_s_index < self.similarity_threshold:
                self.plot_city()
                self.update()
                no_iterations += 1
                
            else:
                print('\n\nOUTCOME OF SIMULATION')
                print(f'It has taken {no_iterations} iterations to reach a citywide similarity threshold of {self.similarity_threshold*100}%')
                self.plot_city()
                break
```

```python
city = Schelling(12,10,0.3,0.5,10,3)
city.run_sim_until_similar()
```
